Data Collection 

```python
df = pd.read_csv('data.csv')
```

```python
df
```

```python
df['MarketingCosts'] = 50
df
```

```python
df['revenue'] = df['Quantity'] * df['Price']
df
```

Adding the retention period necessary for the CLTV dashboard        

```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'], format='mixed', errors='coerce')
df
```

```python
customer_dates = df.groupby('Customer ID')['InvoiceDate'].agg(['min','max'])
```

```python
customer_dates['RetentionPeriod'] = (customer_dates['max'] - customer_dates['min']).dt.days
```

```python
data_modified = df.merge(customer_dates['RetentionPeriod'], left_on='Customer ID', right_index=True)
```

```python
data_modified
```

Data Preparation 

1. Removing Dublicates

```python
data_modified = data_modified.drop_duplicates()
```

2. Handling the missing values

```python
missing = data_modified[data_modified['Customer ID'].isnull()]
```

```python
missing.head()
```

```python
data_modified = data_modified.dropna(subset=['Customer ID'])
```

3. Ensure proper date formats 

```python
data_modified['InvoiceDate'] = pd.to_datetime(data_modified['InvoiceDate'])
```

4. Remove invalid or negative values

```python
data_modified = data_modified[(data_modified['Quantity'] > 0) & (data_modified['Price'] > 0)]
```

5. Aggregate data by customer ID

```python
customer_aggregates = data_modified.groupby('Customer ID').agg(
    TotalRevenue=('revenue', 'sum'),
    TotalOrders=('Invoice', 'count'),
    FirstPurchase=('InvoiceDate', 'min'),
    LastPurchase=('InvoiceDate', 'max')
)
```

```python
customer_aggregates
```

```python
customer_aggregates['RetentionPeriodDays'] = (
    customer_aggregates['LastPurchase'] - customer_aggregates['FirstPurchase']
).dt.days
```

```python
customer_aggregates
```

DATA TRANSFORMATION

```python
data_modified['revenue'] = data_modified['Price']*data_modified['Quantity']
```

```python
customer_aggregates['AverageOrderValue'] = (
    customer_aggregates['TotalRevenue'] / customer_aggregates['TotalOrders']
)
```

```python
customer_aggregates
```

* Segment customers based on revenue

```python
customer_aggregates['RevenueSegment'] = pd.cut(
    customer_aggregates['TotalRevenue'],
    bins=[0, 100, 500, 1000, float('inf')],
    labels=['Low', 'Medium', 'High', 'VIP']
)
```

```python
customer_aggregates
```

* Segment customer based on frequency of their order

```python
customer_aggregates['RevenueSegment'] = pd.cut(
    customer_aggregates['TotalRevenue'],
    bins=[0, 100, 500, 1000, float('inf')],
    labels=['Low', 'Medium', 'High', 'VIP']
)
```

* Segment customer based on region

```python
customer_aggregates['RegionSegment'] = data_modified.groupby('Customer ID')['Country'].first()
```

```python
customer_aggregates
```

CALCULATING Purchase frequency

```python
total_customers = customer_aggregates.shape[0]
```

```python
print(total_customers)
```

```python
purchase_frequency = customer_aggregates['TotalOrders'].sum() / total_customers
print(f"Purchase Frequency: {purchase_frequency}")
```

```python
customer_aggregates['RetentionPeriodYears'] = customer_aggregates['RetentionPeriodDays'] / 365
```

```python
customer_aggregates['PurchaseFrequency'] = (
    customer_aggregates['TotalOrders'] / customer_aggregates['RetentionPeriodYears']
)
```

```python
customer_aggregates['PurchaseFrequency'] = customer_aggregates['PurchaseFrequency'].replace([float('inf'), -float('inf')], 0)
```

```python
customer_aggregates
```

Finally Calculate CLTV

```python
customer_aggregates['CLTV'] = (
    customer_aggregates['AverageOrderValue'] * 
    customer_aggregates['PurchaseFrequency'] * 
    customer_aggregates['RetentionPeriodYears']
)
```

```python
customer_aggregates
```

```python
customer_aggregates['CLTV'] = np.ceil(customer_aggregates['CLTV'])
```

```python
customer_aggregates
```
